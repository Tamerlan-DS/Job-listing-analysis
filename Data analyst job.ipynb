## 4. Data preparation and cleaning

### 4.1 Removing unnecessary columns

```python
jobs.drop(['Unnamed: 0','Headquarters','Easy Apply','Competitors'], axis='columns', inplace=True)
```

### 4.2 Removing faulty rows

```python
jobs = jobs.drop(labels=range(3692, 4092), axis=0)
jobs.shape
```

### 4.3 Casting to suitable data types

```python
jobs = jobs.astype({"index": int,
                    "Job Title": "string",
                    "Job Description": "string",
                    "Company Name": "string",
                    "Salary Estimate": "string",
                    "Rating": float,
                    "Location": "string",
                    "Size": "string",
                    "Founded": int,
                    "Type of ownership": "string", 
                    "Industry": "string",
                    "Sector": "string", 
                    "Revenue": "string"})
jobs.dtypes
```

### 4.4 Extracting necessary data ['Min Salary', 'Max Salary', 'Avg Salary']

```python
jobs['Salary Estimate'] = jobs['Salary Estimate'].str.replace('-','').str.split('K',n=2)
jobs.loc[:, 'Min Salary'] = jobs['Salary Estimate'].map(lambda x: x[0][1:])
jobs.loc[:, 'Max Salary'] = jobs['Salary Estimate'].map(lambda x: x[1][1:])
```

```python
jobs = jobs.astype({"Min Salary": int,
                    "Max Salary": int})
jobs['Min Salary'] = jobs['Min Salary'] * 1000
jobs['Max Salary'] = jobs['Max Salary'] * 1000
jobs['Avg Salary'] = (jobs['Min Salary'] + jobs['Max Salary'])/2
jobs.drop(['Salary Estimate'], axis=1, inplace=True)
jobs['Job Description'] = jobs['Job Description'].str.replace('\n','')
```

### 4.5 Extracting necessary data ['Min Revenue', 'Max Revenue', 'Avg Revenue']

```python
jobs['Revenue'].unique()
```

```python
def get_min_revenue(x_revenue):
    if x_revenue[0] == '$':
        counter_index = 1
        while(x_revenue[counter_index] != ' ' and x_revenue[counter_index] != '+'):
            counter_index = counter_index + 1
        min_revenue = int(x_revenue[1:counter_index])
    else:
        min_revenue = 0
    if (x_revenue.find('million') != -1):
        min_revenue = min_revenue * 1000000
    if (x_revenue.find('billion') != -1):
        min_revenue = min_revenue * 1000000000
    return min_revenue
```

```python
def get_max_revenue(x_revenue):
    x_revenue = x_revenue.replace(' ','')
    if (x_revenue[0] == '$'):
        if (x_revenue.find('+') != -1):
            return 10000000000
        else:
            start = x_revenue.find('to') + 3
            counter_index = start
            while(x_revenue[counter_index] != 'm' and x_revenue[counter_index] != 'b'):
                counter_index = counter_index + 1
        max_revenue = int(x_revenue[start:counter_index])
    elif (x_revenue[0] == 'L'):
        return 1000000
    else:
        max_revenue = 0
    if (x_revenue.find('million') != -1):
        max_revenue = max_revenue * 1000000
    if (x_revenue.find('billion') != -1):
        max_revenue = max_revenue * 1000000000
    return max_revenue
```

```python
jobs['Min Revenue'] = jobs['Revenue'].apply(lambda x : get_min_revenue(x))
jobs['Max Revenue'] = jobs['Revenue'].apply(lambda x : get_max_revenue(x))
jobs['Avg Revenue'] = (jobs['Min Revenue'] + jobs['Max Revenue'])/2
```

# 5.Data analysis and visualization

## Q1) What types of analysts are popular ? 

```python
hr_analysts = jobs[jobs['Job Title'].str.contains('HR')]
engineers = jobs[jobs['Job Title'].str.contains('Engineer')]
sales_analysts = jobs[jobs['Job Title'].str.contains('Sale')]
system_analysts = jobs[jobs['Job Title'].str.contains('System')]
senior_analysts = jobs[jobs['Job Title'].str.contains('Senior')]
finance_analysts = jobs[jobs['Job Title'].str.contains('Finan')]
security_analysts = jobs[jobs['Job Title'].str.contains('Security')]
management_analysts = jobs[jobs['Job Title'].str.contains('Management')]
intelligence_analysts = jobs[jobs['Job Title'].str.contains('Intelligence')]
```

```python
types = [len(hr_analysts),len(engineers),len(sales_analysts),len(system_analysts),
           len(senior_analysts),len(finance_analysts),len(security_analysts),
           len(management_analysts),len(intelligence_analysts)]
type_labels = ['HR Analysts','Data Engineers','Sales Analystst',
             'Systems Analysts','Senior Analysts','Financial Analysts',
             'Security Analysts','Management Analysts','Intelligence Analysts']
plt.figure(figsize=(15,15))
plt.pie(types,labels=type_labels,autopct='%1.1f%%')
plt.title('Most popular types of analysts')
plt.show()
```

##  Q3) Are there any differences between old and new companies in terms of            measured indicators ?

```python
jobs["Founded"].replace({-1: 1700}, inplace=True)
```

```python
old_companies = jobs[jobs['Founded'] < 1990]
new_companies = jobs[jobs['Founded'] > 1989]
new_companies
```

### By average revenues 

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle('Average revenues of old and new companies')
axes[0].hist(old_companies['Avg Revenue'],bins=20,range=(0,10000000000),color='purple')
axes[0].set_xlabel('Average revenue of old companies',fontsize=12)
axes[0].set_ylabel('Density',fontsize=12)
axes[1].hist(new_companies['Avg Revenue'],bins=20,range=(0,10000000000),color='purple')
axes[1].set_xlabel('Average revenue of new companies',fontsize=12)
axes[1].set_ylabel('Density',fontsize=12)
plt.show()
```

### By average salaries 

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle('Average salaries of old and new companies')
axes[0].hist(old_companies['Avg Salary'],bins=50,range=(0,100000),color='pink')
axes[0].set_xlabel('Average salary of old companies',fontsize=12)
axes[0].set_ylabel('Density',fontsize=12)
axes[1].hist(new_companies['Avg Salary'],bins=50,range=(0,100000),color='pink')
axes[1].set_xlabel('Average salary of new companies',fontsize=12)
axes[1].set_ylabel('Density',fontsize=12)
plt.show()
```

### Relation between average salaries and average revenues of companies

### By ratings

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle('Ratings of old and new companies')
axes[0].hist(old_companies['Rating'],bins=50,range=(0,5))
axes[0].set_xlabel('Ratings of old companies',fontsize=12)
axes[0].set_ylabel('Density',fontsize=12)
axes[1].hist(new_companies['Rating'],bins=50,range=(0,5))
axes[1].set_xlabel('Ratings of new companies',fontsize=12)
axes[1].set_ylabel('Density',fontsize=12)
plt.show()
```
